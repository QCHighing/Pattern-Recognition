# 数据评价

$$logloss = - \frac{1}{N} \sum_{i=1}^{N} \sum_{j=1}^{M} y_{i,j} \; log(p_{i,j})  \tag{1}$$

其中$N$代表测试数据集中的商品数量，$i,j$分别是商品和类别的索引。当商品$i$属于类别$j$时，$y_{i,j}=1$，否则为0。其中，$p_{i,j}$是预测的概率。

**注意：** 由于存在自然对数，故规定$p = max(min(p,1-10^{-15}) ,\; 10^{-15})$

```python
def evaluation(label,pred_label):
    num = len(label)
    logloss = 0.0
    for i in range(num):
        p = max(min(pred_label[i][label[i]-1],1-10**(-15)),10**(-15))
        logloss += np.log(p)
    logloss = -1*logloss/num
    return logloss
```
